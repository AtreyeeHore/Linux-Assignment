2.What is the difference between hard link and soft link?
- In Linux, hard links point to the same inode number (file content) as the original file, while soft links point to the original file's name (path). This means that hard links behave like multiple names for the same file, while soft links are more like shortcuts.

Hard Link
- A duplicate reference to the same inode (file data).
- Works only for files, not directories.
 -If the original file is deleted, the data is still accessible via the hard link.
- Both names (original and hard link) are treated equally by the system.

Soft Link (Symbolic Link or Symlink)
- A shortcut that points to the path of another file or directory.
- Can link to directories and across different file systems.
- If the original file is deleted, the symlink becomes broken (dangling link).
- Like a Windows shortcut.

3.What is kernel in Linux?
- Kernel is used to communicate between application and the backend.
- The kernel is the core part of the Linux operating system. It acts as a bridge between the hardware and software. Without the kernel, Linux (or any OS) wouldn’t function.
- It's responsible for managing system resources like memory, CPU, and devices, allowing the OS to effectively interact with the underlying hardware and software components. Essentially, it's the "brain" of the OS, managing how the computer interacts with its resources.
